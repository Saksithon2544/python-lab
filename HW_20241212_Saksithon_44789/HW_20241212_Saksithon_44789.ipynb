```python
import speech_recognition as sr
import pyttsx3
import webbrowser
import datetime
import os
import time

# ฟังก์ชันสำหรับการพูดข้อความ
def speak(text):
    engine.setProperty('rate', 190)
    engine.say(text)
    engine.runAndWait()

# ฟังก์ชันสำหรับการฟังคำสั่งจากไมโครโฟน
def take_command():
    with sr.Microphone() as source:
        recognizer.adjust_for_ambient_noise(source)
        print("กำลังฟังคำสั่ง...")
        try:
            voice = recognizer.listen(source, timeout=10) 
            statement = recognizer.recognize_google(voice, language='th-TH')
            print(f"ผู้ใช้พูดว่า: {statement}")
        except sr.WaitTimeoutError:
            print("หมดเวลาในการฟังคำสั่ง")
            speak("หมดเวลาในการฟังคำสั่ง, กรุณาพูดใหม่อีกครั้ง")
            return "None"
        except Exception:
            print("ขอโทษ, กรุณาพูดใหม่อีกครั้ง")
            speak("ขอโทษ, กรุณาพูดใหม่อีกครั้ง")
            return "None"
        return statement.lower()


# ฟังก์ชันสำหรับการทักทาย
def greeting():
    hour = datetime.datetime.now().hour
    if 0 <= hour < 12:
        speak("สวัสดีตอนเช้า!")
    elif 12 <= hour < 18:
        speak("สวัสดีตอนบ่าย!")
    else:
        speak("สวัสดีตอนเย็น!")
        
# ฟังก์ชันสำหรับการเปิดเว็บไซต์
def open_website(site_name):
    if site_name == "youtube":
        webbrowser.open_new("https://www.youtube.com")
        speak("เปิด YouTube แล้ว")
    elif site_name == "google":
        webbrowser.open_new("https://www.google.com")
        speak("เปิด Google แล้ว")
    else:
        speak("ไม่พบเว็บไซต์นี้")
        
# ฟังก์ชันสำหรับการเปิดโปรแกรม
def open_program(program_name):
    if program_name == "notepad":
        os.system("notepad.exe")
        speak("เปิด Notepad แล้ว")
    elif program_name == "calculator":
        os.system("calc.exe")
        speak("เปิด Calculator แล้ว")
    else:
        speak("ไม่พบโปรแกรมนี้")

# ฟังก์ชันสำหรับการบอกเวลา
def tell_time():
    current_time = datetime.datetime.now().strftime("%H:%M:%S")
    speak(f"ตอนนี้เวลา {current_time}")

# ฟังก์ชันหลักที่ทำงานตลอดเวลา
recognizer = sr.Recognizer()
engine = pyttsx3.init()

# ตั้งค่าเสียงให้เป็นภาษาไทย
voices = engine.getProperty('voices')
for voice in voices:
    if 'thai' in voice.languages:
        engine.setProperty('voice', voice.id)
        break

# โหลดผู้ช่วย AI
print("กำลังโหลดผู้ช่วย AI ส่วนตัวของคุณ - Javis")
speak("กำลังโหลดผู้ช่วย AI ส่วนตัวของคุณ - Javis")

greeting()

while True:
    print("ฉันพร้อมที่จะช่วยคุณ, คุณต้องการให้ฉันทำอะไร?")
    speak("ฉันพร้อมที่จะช่วยคุณ, คุณต้องการให้ฉันทำอะไร?")

    statement = take_command()

    if statement == "None":
        continue
    if "ลาก่อน" in statement or "หยุด" in statement or "บ๊ายบาย" in statement:
        print("ผู้ช่วย AI ส่วนตัวของคุณ Javis กำลังปิดการทำงาน, ลาก่อน")
        speak("ผู้ช่วย AI ส่วนตัวของคุณ Javis กำลังปิดการทำงาน, ลาก่อน")
        break

    # คำสั่งในการเปิดเว็บไซต์
    if "youtube" in statement:
        open_website("youtube")
    elif "google" in statement:
        open_website("google")

    # คำสั่งในการเปิดโปรแกรม
    elif "notepad" in statement:
        open_program("notepad")
    elif "calculator" in statement:
        open_program("calculator")

    # คำสั่งในการบอกเวลา
    elif "เวลา" in statement:
        tell_time()

    # คำสั่งในการถามว่าเป็นยังไงบ้าง
    elif "สบายดีไหม" in statement:
        speak("ฉันสบายดีมาก ขอบคุณที่ถาม!")
        
    # คำสั่งอื่นๆ
    else:
        speak("ขอโทษ, ฉันไม่เข้าใจคำสั่งนั้น")
```
