```python
import os
import pdfplumber
from docx import Document
import yt_dlp as youtube_dl

def download_youtube_video():
    videourl = input("กรุณาใส่ URL ของวิดีโอ YouTube: ")
    path = os.getcwd() 

    ydl_opts = {
        'format': 'best[ext=mp4]',
        'outtmpl': os.path.join(path, '%(title)s.%(ext)s'),
    }
    
    try:
        with youtube_dl.YoutubeDL(ydl_opts) as ydl:
            ydl.download([videourl])
        print(f"ดาวน์โหลดเสร็จสิ้น! วิดีโอถูกบันทึกใน: {path}")
    except Exception as e:
        print(f"เกิดข้อผิดพลาดในการดาวน์โหลดวิดีโอ: {e}")


def convert_pdf_to_docx():
    pdf_file = input("กรุณาใส่ชื่อไฟล์ PDF ที่ต้องการแปลง (เช่น sample.pdf): ")
    docx_path = os.path.join(os.getcwd(), "output_docx.docx") 

    doc = Document()

    try:
        with pdfplumber.open(pdf_file) as pdf:
            for page_number, page in enumerate(pdf.pages):
                text = page.extract_text()
                if text:
                    doc.add_paragraph(text)
                    doc.add_page_break()
                else:
                    print(f"คำเตือน: ไม่มีข้อความในหน้า {page_number + 1}.")

        doc.save(docx_path)
        print(f"บันทึกไฟล์เรียบร้อยแล้ว: {docx_path}")
    except Exception as e:
        print(f"เกิดข้อผิดพลาดในการประมวลผล PDF: {e}")

def main():
    while True:
        print("\n--- เมนูโปรแกรม ---")
        print("1. ดาวน์โหลดวิดีโอ YouTube")
        print("2. แปลง PDF เป็นเอกสาร Word")
        print("3. ออกจากโปรแกรม")

        choice = input("กรุณาเลือก (1-3): ")

        if choice == "1":
            download_youtube_video()
        elif choice == "2":
            convert_pdf_to_docx()
        elif choice == "3":
            print("ออกจากโปรแกรม ขอบคุณที่ใช้งาน!")
            break
        else:
            print("ตัวเลือกไม่ถูกต้อง กรุณาลองใหม่อีกครั้ง")

if __name__ == "__main__":
    main()
```
